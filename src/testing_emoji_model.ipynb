```python
from emoji import UNICODE_EMOJI
def is_itemoji(s, language="en"):
    return s in UNICODE_EMOJI[language]
```

```python
def getfirstword(lst):
    for i in lst:
        if is_itemoji(i)==False:
          return i  
 
    
```

```python
word=getfirstword(lst)
print(word)
```
